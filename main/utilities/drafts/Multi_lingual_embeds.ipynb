```python
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
# Load a multilingual embedding model
# model = SentenceTransformer("all-MiniLM-L6-v2")
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
# Define Persian and English equivalent sentences
persian_text = "من امروز به کتابخانه رفتم."
english_text = "I went to the library today."

persian_text = "beautiful"
english_text = "ugly"

persian_text = "من امروز به کتابخانه رفتم."
english_text = "A Comprehensive Introduction to Differential Geometry"

# persian_text = "I am happy today."  # "I am happy today."  , "من امروز خوشحال هستم."
# english_text = "you am very sad today."

# Generate embeddings
persian_embedding = model.encode(persian_text, convert_to_tensor=True)
english_embedding = model.encode(english_text, convert_to_tensor=True)

# Compute cosine similarity
similarity = util.pytorch_cos_sim(persian_embedding, english_embedding)

# Print similarity score
print(f"Cosine Similarity: {similarity.item():.4f}")

# Interpretation:
# - Closer to 1 → Very similar
# - Around 0 → Unrelated
# - Negative → Opposite meaning
```
